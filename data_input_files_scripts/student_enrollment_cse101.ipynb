##### Reading OBE Marksheet

```python
import pandas as pd
file = pd.read_excel('CSE101 OBE Mark Sheet.xlsx')
```

```python
file.head()
```

##### Slicing data for student enrollment table

```python
enrollment = file.iloc[3:, :5]   # taking upto semester column 
enrollment.head()
```

```python
enrollment = enrollment.drop(enrollment.columns[[2]], axis= 1) # dropping course ID column # As we need 4 columns
enrollment.head() 
```

```python
# Converting dataframe to list
enrollment_l = enrollment.values.tolist()
```

```python
for i in range(len(enrollment_l)):
    enrollment_l[i][0] = 1000 + i
    enrollment_l[i][1] = int(enrollment_l[i][1])
    enrollment_l[i][2] = 'cse101_' + str(int(enrollment_l[i][2]))
    
print(enrollment_l[13])
```

```python
# Converting list dataframe
df = pd.DataFrame(enrollment_l,columns=['Student Enrollment ID', 'Student ID', 'Section ID', 'Semester'] )
df.head()
```

```python
df.to_excel('student_enrollment_cse101.xlsx')
```
